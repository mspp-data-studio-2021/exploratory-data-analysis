```python
pluto_raw = pd.read_csv("data-raw/pluto_20v4.csv", 
                        dtype = {"ltdheight" : "object", "zonedist4" : "object",
                                 "zonedist2" : "object", "overlay1" :  "object",
                                 "spdist1" : "object", "histdist" :  "object",
                                 "version" : "object"})
```

```python
pluto_raw.info()
```

```python
pluto_raw.describe()
```

```python
pluto_hotels = (pluto_raw
    .query("bldgclass.str.match(r'H') > 0")
    .loc[:, ['bbl', 'borough', 'bldgarea', 'latitude', 'longitude']]
)

pluto_hotels
```

```python
boro_hotel_stats = (pluto_hotels
 .groupby(['borough'])
 .agg(
    n_hotels=('bbl', 'count'), 
    tot_bldg_area=('bldgarea', 'sum')
 )
 .reset_index()
 .head()
)

boro_hotel_stats
```

```python
import altair as alt

alt.Chart(boro_hotel_stats).mark_bar().encode(
  x='borough',
  y='n_hotels'
)
```
